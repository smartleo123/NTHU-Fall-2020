```
import tensorflow as tf
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```
mp1=100 # 100 lines
np1=50 # each line with 30 points
noise1 = (2.0 * np.random.random((mp1,np1)) - 1) * 0.1
X_train1 = np.ones((mp1,np1)) + noise1
print(X_train1)
```

```
X_train2 = np.ones((mp1, np1))
for row in X_train2:
    randnum = int(np.random.random(1) * 45 + 1) # the key point : )
    row[randnum] = 0.1
    row[randnum + 1] = 0.1
    row[randnum + 2] = 0.1
    row[randnum + 3] = 0.1
    row[randnum + 4] = 0.1

print(randnum)

X_train2 = X_train2 + noise1
print(X_train2)
```

```
Y_train1 = np.zeros(mp1)[:,np.newaxis]
Y_train2 = np.ones(mp1)[:,np.newaxis]

X_train = np.concatenate((X_train1, X_train2), axis=0)
Y_train = np.concatenate((Y_train1, Y_train2), axis=0)
X_train = X_train -1
X_train = np.expand_dims(X_train, axis=2)
```

```
from keras.models import Sequential
from keras.layers import Dense, Activation, MaxPooling1D, GlobalAveragePooling1D, Conv1D

model = Sequential()
model.add(Conv1D(32, 3, activation='relu',input_shape=(np1, 1)))
model.add(MaxPooling1D(3))
model.add(Conv1D(16, 3, activation='relu'))
model.add(GlobalAveragePooling1D())
##model.add(Dropout(0.75))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, Y_train, batch_size=32, epochs=10)
loss,accuracy = model.evaluate(X_train, Y_train,batch_size=32)
print ("test loss for testdata: {}".format(loss))
print ("test accuracy for testdata: {}".format(accuracy))
```
