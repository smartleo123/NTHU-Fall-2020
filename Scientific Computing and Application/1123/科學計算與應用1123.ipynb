## HW2

```
XOR_X = [[0,0],[0,1],[1,0],[1,1]]
XOR_Y = [[0],[1],[1],[0]]
x = tf.placeholder(tf.float32, shape=[4,2])
y = tf.placeholder(tf.float32, shape=[4,1])
w1 = tf.Variable(tf.random_uniform([2,2], -1, 1))
w2 = tf.Variable(tf.random_uniform([2,1], -1, 1))
b1 = tf.Variable(tf.zeros([2]))
b2 = tf.Variable(tf.zeros([1]))
```

```
z1 = tf.nn.relu(tf.matmul(x, w1) + b1)
pred = tf.sigmoid(tf.matmul(z1,w2) + b2)
cost = tf.reduce_mean((tf.nn.sigmoid_cross_entropy_with_logits(labels = y,logits = pred)))
learning_rate = 0.1

optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train_op = optimizer.minimize(cost)
```

```
x_train = np.array(XOR_X)          
y_train = np.array(XOR_Y)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
sess.run(train_op, {x:x_train, y:y_train})
curr_W, curr_b, curr_loss = sess.run([w1, b1, cost], {x:x_train, y:y_train})
print(curr_loss)

epochs=1000
# # 最佳化迴圈

for i in range(epochs):
  sess.run(train_op,{x:x_train, y:y_train})
  curr_W,curr_b,curr_loss = sess.run([w1, b1, cost], {x:x_train, y:y_train})

# 輸出結果
curr_W, curr_b, curr_loss = sess.run([w1, b1, cost], {x:x_train, y:y_train})
print("W: b:  loss: ",curr_W,curr_b,curr_loss)
```

```
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
train_X, train_Y,test_X, test_Y = mnist.train.images, mnist.train.labels, mnist.test.images, mnist.test.labels
```

```
# train_X = tf.placeholder(tf.float32)
# train_Y = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32, shape=[55000,784])
y = tf.placeholder(tf.float32, shape=[55000,10])
w1 = tf.Variable(tf.random_uniform([784,300], -1, 1))
w2 = tf.Variable(tf.random_uniform([300,10], -1, 1))
b1 = tf.Variable(tf.zeros([300]))
b2 = tf.Variable(tf.zeros([10]))

z1 = tf.nn.relu(tf.matmul(x, w1) + b1)
z2 = tf.matmul(z1,w2) + b2
```

```
cost = tf.reduce_mean((tf.nn.softmax_cross_entropy_with_logits(labels = y,logits = z2)))
learning_rate = 0.01

optimizer = tf.train.GradientDescentOptimizer(learning_rate)
train_op = optimizer.minimize(cost)
```

```
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

epochs=10

# # 最佳化迴圈

for i in range(epochs):
  sess.run(train_op,{x:train_X, y:train_Y})
  curr_W,curr_b,curr_loss = sess.run([w1, b1, cost], {x:train_X, y:train_Y})

# 輸出結果
curr_W, curr_b, curr_loss = sess.run([w1, b1, cost], {x:train_X, y:train_Y})
cost = curr_loss
print('cost value',cost)
```
