## HW1

```
import matplotlib.pyplot as plt
TRAINING_STEPS=10
x=tf.Variable(tf.constant(9,dtype=tf.float32),name='x')
y=tf.square(x)
train_op=tf.train.GradientDescentOptimizer(0.2).minimize(y)
#實現梯度下降算法的優化器
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(TRAINING_STEPS):
        sess.run(train_op)
        x_value=sess.run(x)
        print(i+1, x_value)
```

```
x = tf.Variable(tf.constant(-100,dtype=tf.float32),name='x')
y = tf.Variable(tf.constant(100,dtype=tf.float32),name='y')
z = 0.5*tf.pow(x,2) + 2.5*tf.pow(y,2)
```

```
train_step1 = tf.train.GradientDescentOptimizer(learning_rate = 0.35).minimize(z)
TRAINING_STEPS=30
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(TRAINING_STEPS):
        sess.run(train_step1)
        x_value=sess.run(x)
        y_value=sess.run(y)
        print(i+1, x_value,y_value)
```

```
train_step1 = tf.train.MomentumOptimizer(learning_rate = 0.03, momentum=0.9).minimize(z)
TRAINING_STEPS=50
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(TRAINING_STEPS):
        sess.run(train_step1)
        x_value=sess.run(x)
        y_value=sess.run(y)
        print(i+1, x_value,y_value)
```
