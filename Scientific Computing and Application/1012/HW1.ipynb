```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
x = np.linspace(-1,1,100)
signal = 2 + x + 2 * x * x
noise = np.random.normal(0, 0.2, 100)
y = signal + noise

#####################################

xtrain = x[0:80]
ytrain = y[0:80]

one = np.ones([80,1])

xtrain2 = np.square(xtrain)
xtrain3 = np.power(xtrain, 3)
h=np.column_stack((one,xtrain,xtrain2))
print(h)

h_trans = h.transpose()

temp = np.dot(h_trans, h)

inverse = np.linalg.inv(temp)
temp2 =np.dot(h_trans,ytrain)

A = np.dot(inverse,temp2)

print(A)

f_train=np.dot(h,A)
print(f_train)


least_square = ytrain - f_train
least_square_transpose = least_square.transpose()

np.dot(least_square_transpose, least_square)
error = np.sqrt(np.sum(np.dot(least_square_transpose, least_square)))
print(error)
```
